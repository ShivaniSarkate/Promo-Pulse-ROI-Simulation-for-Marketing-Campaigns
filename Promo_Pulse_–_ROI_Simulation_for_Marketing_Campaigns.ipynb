**Project Title : PromoPulse - ROI Simulation for Marketing Campaigns**

**Objective** : Simulate the financial impact of a marketing campaign by modeling conversion rates, cost per acquisition, and ROI across budget scenarios.

**Business Context** : Marketing teams need to justify spend with measurable impact. ROI modeling helps forecast performance, optimize budgets, and prioritize high-return channels.

**Workflow Overview** :
- Simulate campaign spend and conversion rates  
- Calculate cost per acquisition (CPA) and ROI  
- Visualize ROI vs. spend curve  
- Identify breakeven point  
- Recommend optimal budget allocation

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Simulate campaign scenarios
np.random.seed(42)
budgets = np.linspace(10000, 100000, 20)
conversion_rate = 0.05 + np.random.normal(0, 0.005, len(budgets))
avg_order_value = 1200
cost_per_click = 20
clicks = budgets / cost_per_click
conversions = clicks * conversion_rate
revenue = conversions * avg_order_value
roi = (revenue - budgets) / budgets

df = pd.DataFrame({
    'Budget': budgets,
    'Conversions': conversions,
    'Revenue': revenue,
    'ROI': roi
})

# Step 3: Visualize ROI curve
plt.figure(figsize=(10,6))
plt.plot(df['Budget'], df['ROI'], marker='o', color='green')
plt.axhline(0, color='red', linestyle='--')
plt.title('ROI vs. Campaign Budget')
plt.xlabel('Budget (â‚¹)')
plt.ylabel('ROI')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 4: Breakeven point
breakeven = df[df['ROI'] > 0].iloc[0]['Budget']
print(f"ðŸ’¡ Breakeven Budget: â‚¹{breakeven:.0f}")
```
